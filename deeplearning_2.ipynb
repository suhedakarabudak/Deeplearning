-Swish fonksiyonu


```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
#Linear Function
def linear(x,m,b):
    return x * m + b

x= np.linspace(0,10,100)
y=linear(x,3,6)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function')
plt.show()

```

```python
#Sigmoid Function

def sigmoid(x):
    return 1 / (1+ np.exp(-x))

x=np.linspace(-10,10,100)

y=sigmoid(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid function')
plt.show()

```

```python
#Hiperbolik Tanjant Function

def hiperbolik_tan(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

x=np.linspace(-10,10,100)

y=hiperbolik_tan(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hiperbolik Tanjant function')
plt.show()

```

```python
#Relu function

def relu(x):
    return np.maximum(0,x)


x=np.linspace(-10,10,100)

y=relu(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relu function')
plt.show()

```

```python
#X değerleri aralığı
x = np.linspace(0, 2*np.pi, 100)

y1 = linear(x,3,6)
y2 = sigmoid(x)
y3 = hiperbolik_tan(x)
y4 = relu(x)

# Grafik çizimi
plt.plot(x, y1, label='Linear')
plt.plot(x, y2, label='Sigmoid')
plt.plot(x, y3, label='Hiperbolik_tan')
plt.plot(x, y4, label='Relu')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Activation Function Graph')
plt.legend() 
plt.show()
```
